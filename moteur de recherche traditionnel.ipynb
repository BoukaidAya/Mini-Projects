```python
import wikipediaapi

# Define your user-agent
USER_AGENT = "MyWikipediaBot/1.0 (https://example.com; myemail@example.com)"

search_term = input("Entrez un terme de recherche: ")

# Create a Wikipedia object with the specified user-agent
wiki_wiki = wikipediaapi.Wikipedia(
    language='fr',
    user_agent=USER_AGENT
)

page = wiki_wiki.page(search_term)

if page.exists():
    print("Titre :", page.title)
    print("Contenu :", page.text)  
else:
    print("Aucun résultat trouvé")
```
