```python
from scipy.io.wavfile import write
```

```python
fs = 44100 #frequence 

sec = int(input("Temps du rec : (seconde)")) #Temps enregistrement

print("Rec en cours ...")

record = sounddevice.rec(int(sec * fs), samplerate=fs,channels=2)

sounddevice.wait()

write("son.wav",fs,record)

print("fin")
```
